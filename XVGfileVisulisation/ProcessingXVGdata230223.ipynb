## Tranform xvg files to excel/csv files

**Note, the commands in the two cells below only need to be processed once, at the start. After that, since all the excel and csv files have already been generated, there is no need to run the commands again.**

```python
# save as csv files

import pandas as pd
import os

rmsd = 'rmsd'
rmsf = 'rmsf'
gyrate = 'gyrate'

mypath = 'XVGfile6replicas230223/' # change it to the directory where your xvg files are
newpath = 'CSVFileGeneratedFromXVG230223/' # the new directory where you want to save the excel files

directory = os.fsencode(mypath)


# iterate over files in the given directory
# reference: https://stackoverflow.com/questions/10377998/how-can-i-iterate-over-files-in-a-given-directory
# RMSD generator
    
for file in os.listdir(directory):
    
    fname = os.fsdecode(file)
    
    # according to Chengyu, it is better to change the path name using os.path.join
    # apath = 'ExtractXVG280522test2/'+ fname
    apath = os.path.join(mypath,fname)
    xpath = os.path.join(newpath,fname)
    
    # to check if the xvg file is rmsd
    # reference: https://stackoverflow.com/questions/33743816/how-to-find-a-filename-that-contains-a-given-string
    if rmsd in fname: 
        with open(apath,'r') as f:
            data = []
            for line in f:
                if not line.startswith("#"):
                    if not line.startswith('@'):
                        new_list = [elem for elem in line.split()]
                        data.append(new_list)
        df = pd.DataFrame (data, columns = ['Time', 'RMSD'])
        path = os.path.splitext(xpath)[0]
        df.to_csv(path + ".csv", index = False) 
    elif rmsf in fname: 
        with open(apath,'r') as f:
            data = []
            for line in f:
                if not line.startswith("#"):
                    if not line.startswith('@'):
                        new_list = [elem for elem in line.split()]
                        data.append(new_list)
        df = pd.DataFrame (data, columns = ['Residue', 'RMSF'])
        path = os.path.splitext(xpath)[0]
        df.to_csv(path + ".csv", index = False) 
                               
    elif gyrate in fname: 
        with open(apath,'r') as f:
            data = []
            for line in f:
                if not line.startswith("#"):
                    if not line.startswith('@'):
                        new_list = [elem for elem in line.split()]
                        data.append(new_list)
        df = pd.DataFrame (data, columns = ['Time', 'Rg','x','y','z'])
        df_f = df[["Time",'Rg']]
        path = os.path.splitext(xpath)[0]
        df_f.to_csv(path + ".csv", index = False) 
                             
    else:
         continue
```

Generate dictionaries to store rmsd, rmsf, and gyrate values

```python
# reference: https://stackoverflow.com/questions/55245721/read-in-multiple-csv-into-separate-dataframes-in-pandas
# https://stackoverflow.com/questions/36294712/python-filename-contains-string-metachar
# https://www.delftstack.com/howto/python/python-fnmatch/

# save rmsd, rmsf, gyrate files into different dictionaries

import os
import pandas as pd
import fnmatch # looking for a specific string in a file name

directory_csv = csvpath

# csv rmsd files
csvs_rmsd = [x for x in os.listdir(directory_csv) if fnmatch.fnmatch(x, '*rmsd*.csv')] # fnmatch look for rmsd filename specifically

# rmsf
csvs_rmsf = [x for x in os.listdir(directory_csv) if fnmatch.fnmatch(x, '*rmsf*.csv')] 

# gyrate
csvs_gyrate = [x for x in os.listdir(directory_csv) if fnmatch.fnmatch(x, '*gyrate*.csv')] 

## example for fnmatch: the files that start with hello and end with .js `if fnmatch.fnmatch(file, "hello*.js"):`

d_rmsd = {}  # dictionary that will hold them 
d_rmsf = {}
d_gyrate = {}

for file_name in csvs_rmsd:  # loop over files

   # read csv into a dataframe and add it to dict with file_name as it key
   d_rmsd[file_name] = pd.read_csv(os.path.join(directory_csv,file_name)) # to note, directory_csv and file_name need to be joint or else the file can't be found
    
for file_name in csvs_rmsf:  # loop over files

   # read csv into a dataframe and add it to dict with file_name as it key
   d_rmsf[file_name] = pd.read_csv(os.path.join(directory_csv,file_name))

for file_name in csvs_gyrate:  # loop over files

   # read csv into a dataframe and add it to dict with file_name as it key
   d_gyrate[file_name] = pd.read_csv(os.path.join(directory_csv,file_name))
```

# Pre-processing for Rsq analysis

## Merge all the RMSF results together into one dataframe

```python
number = []
for key in d_rmsf:
    print(key)
    number.append(key)
print(len(number))
```

```python
# To merge all the RMSF results togehter into one dataframe
# After running this command, we should be able to group by condition or residue
import pandas as pd

result = {}

for i in condition34_clean:
    frames = list()
    
    for key in d_rmsf:
        if re.search(i, key): # note, here is different from the script before. Edited date 201222
            
            frames.append(d_rmsf[key]) 
            
    merge = pd.concat(frames) # merge dataframes which are in the same conditon (i.e. 6 replicas) together
    merge = merge.reset_index(drop=True) #reset index
    merge = merge.assign(Condition = i) # generate a new column called Condition which can be useful in future when combining different temp dataframes and plot distribution
    result[i] = merge # save into the result dictionary

# combine all the dataframes in result together

frames1 = list()
result_comb ={}

number = [] # just to check if the total number of keys is 34

for key in result:
    
    print(result[key])
    frames1.append(result[key])
    print(key) #check key number 
    number.append(key) # check key number
    print(len(number)) # check key number
    #print(frames1)
    
result_comb = pd.concat(frames1) # merge dataframes
result_comb = result_comb.reset_index(drop=True) # reset index, or else there will be errors in the output when plotting
```

## Mean RMSF value of the 6 replicas

```python
# calculate RMSF mean value group by Residue in each condtion
# use groupby function in pandas
# reference: https://jamesrledoux.com/code/group-by-aggregate-pandas

grouped_result_comb = result_comb.groupby(['Condition','Residue']).agg({'RMSF': ['mean']})
grouped_result_comb.columns = ['RMSF_mean']
grouped_result_comb = grouped_result_comb.reset_index()
grouped_result_comb
```

## Build a dataframe with columns as different conditions

```python
# Generate a table as the starting point

RMSF_rsq = d_rmsf['4_5_338_200_r3_rmsf.csv'] # randomly pick up one. Doesn't matter since we just need the first column that has the residue info
RMSF_rsq = RMSF_rsq.drop(columns = ['RMSF'])
RMSF_rsq
```

```python
# drop residue column

RMSF_mean = grouped_result_comb.drop(columns = ['Residue'])
RMSF_mean

# restyle the dataframe
for i in condition34_clean:
    
    filter = RMSF_mean[RMSF_mean["Condition"] == i] # filter by row: https://pandas.pydata.org/docs/getting_started/intro_tutorials/03_subset_data.html
    extracted_col = filter["RMSF_mean"]
    extracted_col = extracted_col.reset_index(drop=True) # you need to reset index, or else the indexes of the dataframe and the columns (the second one and those afterwards) to be added are not exactly equal, NaNs will result: https://stackoverflow.com/questions/26221300/nan-values-when-new-column-added-to-pandas-dataframe 
    print("column to added:")
    display(extracted_col)
            
    RMSF_rsq.insert(1, i, extracted_col)
    print("RMSF_rsq after adding column:")
    display(RMSF_rsq)
```
